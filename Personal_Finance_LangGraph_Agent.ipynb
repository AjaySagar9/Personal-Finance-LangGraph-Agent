```python
from typing import TypedDict

class FinanceState(TypedDict):

    salary: float

    food: float

    rent: float

    travel: float

    shopping: float

    total_expense: float

    savings: float

    advice: str
```

```python
def expense_agent(state):

    total = (

        state["food"]

        + state["rent"]

        + state["travel"]

        + state["shopping"]

    )

    return {

        "total_expense": total

    }
```

```python
def savings_agent(state):

    savings = (

        state["salary"]

        - state["total_expense"]

    )

    return {

        "savings": savings

    }
```

```python
def advice_agent(state):

    savings = state["savings"]

    salary = state["salary"]

    savings_percent = (

        savings / salary

    ) * 100

    if savings_percent >= 30:

        advice = (
            "Excellent Savings Habit"
        )

    elif savings_percent >= 15:

        advice = (
            "Good Savings Habit"
        )

    else:

        advice = (
            "Reduce Expenses and Save More"
        )

    return {

        "advice": advice

    }
```

```python
def report_agent(state):

    print("\n------ FINANCE REPORT ------")

    print(
        "Salary:",
        state["salary"]
    )

    print(
        "Total Expense:",
        state["total_expense"]
    )

    print(
        "Savings:",
        state["savings"]
    )

    print(
        "Advice:",
        state["advice"]
    )

    return {}
```

```python
from langgraph.graph import StateGraph

builder = StateGraph(
    FinanceState
)

builder.add_node(
    "expense",
    expense_agent
)

builder.add_node(
    "savings",
    savings_agent
)

builder.add_node(
    "advice",
    advice_agent
)

builder.add_node(
    "report",
    report_agent
)

builder.set_entry_point(
    "expense"
)

builder.add_edge(
    "expense",
    "savings"
)

builder.add_edge(
    "savings",
    "advice"
)

builder.add_edge(
    "advice",
    "report"
)

graph = builder.compile()
```

```python
salary = float(
    input("Salary: ")
)

food = float(
    input("Food Expense: ")
)

rent = float(
    input("Rent Expense: ")
)

travel = float(
    input("Travel Expense: ")
)

shopping = float(
    input("Shopping Expense: ")
)

result = graph.invoke({

    "salary": salary,

    "food": food,

    "rent": rent,

    "travel": travel,

    "shopping": shopping

})
```
